```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
### matplotlib partの寄せ集めのイメージ
```

```python
fig = plt.figure(figsize=(5,5))
ax = plt.axes()
```

```python
#### 折線グラフ ax.plot(x,y)
```

```python
x = [105, 210, 301, 440, 580]
y = [18, 20 ,30, 50, 60]

fig = plt.figure(figsize=(5,5))
ax = plt.axes()

ax.plot(x,y)
plt.show()
```

```python
### タイトル、軸名の付け方
```

```python
x = [105, 210, 301, 440, 580]
y = [18, 20 ,30, 50, 60]

fig = plt.figure(figsize=(5,5))
ax = plt.axes()

ax.set_title('practice_matplotlib')
ax.set_xlabel('time')
ax.set_ylabel('distance')
ax.grid()

ax.plot(x,y)
plt.show()
```

```python
### グラフの種類や色の指定
```

```python
x = [105, 210, 301, 440, 580]
y = [18, 20 ,30, 50, 60]

fig = plt.figure(figsize=(5,5))
ax = plt.axes()

ax.set_title('practice_matplotlib')
ax.set_xlabel('time')
ax.set_ylabel('distance')
ax.grid()

ax.plot(x, y, marker="o", color="red", linestyle='--', linewidth = 2.0)
plt.show()
```

```python


### 棒グラフ　ax.bar(x, y)
```

```python
x = ["A", "B", "C", "D"]
y = [400, 300, 150, 300]

fig = plt.figure(figsize=(5,5))
ax = plt.axes()

ax.set_title('practice_matplotlib')

ax.bar(x, y, color="red")
plt.show()
```

```python
### ヒストグラム ax.hist(x)
### numpyで乱数データ使用
```

```python
x = np.random.randn(1000)

fig = plt.figure(figsize=(5,5))
ax = plt.axes()

ax.hist(x, bins=50)
plt.show()
```

```python
### sanpuzu
```

```python
x = np.random.randint(1, 101, 200)
y = np.random.randint(1, 101, 200)

fig = plt.figure(figsize=(5,5))
ax = plt.axes()

ax.scatter(x, y)
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

#グラフの間隔を空ける
plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

```python
fig = plt.figure(figsize=(10,18))

ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

plt.subplots_adjust(wspace=0.5, hspace=0.5)

#ax1
ax1.set_title('ax1_plot')
x = [2, 4, 6, 8, 10]
y = [30, 56, 79, 90, 113] 
ax1.plot(x, y)

#ax2
ax2.set_title('ax2_bar')
x = [2, 4, 6, 8, 10]
y = [30, 56, 79, 90, 113] 
ax2.bar(x, y)

#ax3
ax3.set_title('ax3_hist')
x = np.random.randn(1000)
ax3.hist(x, bins=50)

#ax4
ax4.set_title('ax4_scatter')
x = np.random.randint(1, 101, 200)
y = np.random.randint(1, 101, 200)
ax4.scatter(x, y)
```
