## GROUPBYで複数項目を括る

```python
df_temp3.groupby(['都道府県名', '産業大分類名'])[['所定内給与額（万円）', '年間賞与その他特別給与額（万円）', '一人当たり賃金（万円）']].mean()
```

```python
df_temp3.dtypes
```

```python
df_pref_category['所定内給与額（万円）']= df_pref_category['所定内給与額（万円）'].replace('-', np.nan)
df_pref_category['年間賞与その他特別給与額（万円）']= df_pref_category['年間賞与その他特別給与額（万円）'].replace('-', np.nan)
df_pref_category['一人当たり賃金（万円）']= df_pref_category['一人当たり賃金（万円）'].replace('-', np.nan)
```

```python
df_pref_category.dropna(
    subset=['所定内給与額（万円）'],
    axis=0,
    inplace=True
)
print(df_pref_category.shape)
```

```python
df_pref_category.isnull().sum()
```

```python
df_temp3.dtypes
```

```python
df_pref_category = df_pref_category.astype({'所定内給与額（万円）': float, '年間賞与その他特別給与額（万円）': float, '一人当たり賃金（万円）': float})
print('======データ型\n', df_pref_category.dtypes)
```

### 箱ヒゲ図

年齢階層別の一人当たり賃金を箱ヒゲ図で表示する

```python
print(df_pref_category.shape)
df_pref_category.head()
```

横軸：年齢階級<br>
縦軸：一人当たり賃金

```python
age_list = df_pref_category['年齢'].unique()
print(age_list)
```

```python
wage_list = []
for age in age_list:
    print(age)
    wage_temp = df_pref_category[df_pref_category['年齢'] == age]['一人当たり賃金（万円）'].values.tolist()
    wage_list.append(wage_temp)
```

```python
len(wage_list)
```

```python
len(wage_list[0])
```

```python
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title('年齢階級毎の一人当たり賃金', fontproperties=font_property)
ax.set_xticklabels(age_list, fontproperties=font_property)
ax.boxplot(wage_list)
plt.show()
```

外れ値を無視する<br>
boxplotの showfliers = False

```python
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title('年齢階級毎の一人当たり賃金', fontproperties=font_property)
ax.set_xticklabels(age_list, fontproperties=font_property)
ax.boxplot(wage_list, showfliers = False)
plt.show()
```

### 産業別の一人当たり賃金を箱ヒゲ図表示

```python
print(df_pref_category.shape)
df_pref_category.head()
```

```python
# 産業大分類名をユニークキーにする
ind_list = df_pref_category['産業大分類名'].unique()
print(ind_list)
```

```python
# 産業毎に賃金を抽出する
wage_list_ind = []
for name in ind_list:
    print(name)
    wage_temp_ind = df_pref_category[df_pref_category['産業大分類名'] == name]['一人当たり賃金（万円）'].values.tolist()
    wage_list_ind.append(wage_temp_ind)
```

```python
# 箱ヒゲ図をかく
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title('産業別一人あたり賃金（万円）', fontproperties=font_property)
ax.set_xticklabels(ind_list, fontproperties=font_property, rotation=75) # 横軸の名前が長いので回転
ax.boxplot(wage_list_ind)
plt.show()
```

```python
# 外れ値を除く
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title('産業別一人あたり賃金（万円）', fontproperties=font_property)
ax.set_xticklabels(ind_list, fontproperties=font_property, rotation=75) # 横軸の名前が長いので回転
ax.boxplot(wage_list_ind, showfliers=False)
plt.show()
```
